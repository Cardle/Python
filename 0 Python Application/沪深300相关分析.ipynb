#### 另外一个获取历史数据的函数get_h_data()
在不指定开始时间和结束时间时，该函数默认返回最近一年的日线数据，返回的数据与get_hist_data不同的是，该函数只返回开盘价（open）、最高价（high）、收盘价（close）、最低价（low）、成交量（volume）、成交金额（amount）六列 ，同样加上时间段也可以获取相应数据。

```python
start = '2019-01-01'
end = '2019-04-30'
```

```python
# 获取最新的实时股票状态
data = ts.get_today_all()
data.head()
```

```python
# 选取当日code，name，以及settlement
code_name = data[['code','name','settlement']]
code_name.columns = ['code','name','price']
code_name.head()
```

```python
#提取指定代码和股票名称
code = list(set(code_name['code'].tolist()))
name = list(set(code_name['name'].tolist()))
```

```python
# 预览ts.get_k_data()的数据相关情况
ts.get_k_data('sh',start= start,end=end).head()
```

```python
# 获取时间范围内的所有股票数据
all_data = {ticker:ts.get_k_data(code = ticker,
                                 start = start,
                                 end=end) for ticker in code}
```

```python
# 获取收盘价格
price = pd.DataFrame({ticker:data['close'] for ticker,data in all_data.items()})
price.columns = name
price.head()
```

```python
# 计算收盘价格增幅
pct_price = price.pct_change(fill_method='pad')
pct_price.dropna(how='all',inplace=True)
pct_price.head()
```

```python
result_corr = pd.DataFrame(pct_price.corrwith(pct_price.上证指数),columns=['相关系数']).sort_values(by='相关系数',ascending=False)
result_corr.head(20)                           
```
