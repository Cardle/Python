### 绘制常规柱形图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
%matplotlib inline
```

```python
plt.rcParams['font.sans-serif']='Simhei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
```

```python
x = np.array(['东区','南区','西区','北区'])
y1 = np.array([1000,3000,2000,4000])
y2 = np.array(random.sample(range(3000,8000),4))
print(x)
print(y1)
print(y2)
```

```python
#绘制常规柱形图
plt.figure(figsize=(5,4))
plt.subplot(111)
plt.bar(x,y1,width=0.5,align='center',label='任务量',color=(0.004,0.722,0.667))
plt.title('全国分区任务量',loc='center')
plt.xlabel('区域')
plt.ylabel('任务量')
for a,b in zip(x,y1):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
plt.legend()
```
