### 柱形图

```python
x = np.array(list(range(1,9)))
x
```

```python
y = np.random.randint(1,10,8)
y
```

```python
plt.bar(x,y,width=-0.8,align='edge',
        edgecolor='k',lw=2,color='c',
        ecolor='k',log=False,capsize=10,xerr=0.5,yerr=0.5,
        tick_label=list('qacerjbp'),hatch='//////',
       )
plt.xlabel('箱子编号')
plt.ylabel('箱子重量')
```

### 条形图

```python
plt.barh(x,y,align='center',height=0.8,
        edgecolor='k',lw=2,color='c',
        ecolor='k',log=False,capsize=10,xerr=0.5,yerr=0.5,
        tick_label=list('qacerjbp'),hatch='\\',capstyle='round',
       )
plt.xlabel('箱子编号')
plt.ylabel('箱子重量')
```

### 直方图

```python
plt.figure(figsize=(10,4))
x = np.random.randn(10000)
plt.hist(x,bins='auto',color='#607c8e',align='mid',density=False,stacked=True,cumulative=False,rwidth=0.7)
```

### 饼图

```python
plt.figure(figsize=(10,6))
kinds=['简易箱','保温箱','行李箱','密封箱']
color=['g','b']
soldNums=[50,450,150,350]
explode=[0.0,0.0,0.1,0.0]

a = plt.pie(soldNums,
        radius=1,
        rotatelabels=False,
        frame=False,
        explode=explode,
        pctdistance=.6,
       labels=kinds,
       autopct=lambda x:'{:.2%}'.format(x/100),
                   startangle = 0)

plt.title("不同箱子的销售数量占比")
plt.legend(a[0],labels=kinds,loc='best')

plt.setp(a[1],size=11,weight='bold',color='k')


# 绘制对应表格
cellText = soldNums
colColours = list('cmyk')

plt.table(cellText=[cellText],colWidths=[0.1]*4,colLabels=kinds,cellLoc='center',rowLabels=['Sales'],loc='bottom',
         cellColours=[list('cmyy')])

```

```python
plt.subplot(111,polar=True)
angles = np.linspace(0,2*np.pi,5,endpoint=False)
labels = ['沟通能力','业务理解','逻辑思维','快速学习','工具使用']
data = [2,3.5,4,4.5,5]
data = np.concatenate((data,[data[0]]))
angles = np.concatenate((angles,[angles[0]]))
print(data)
print(angles)

plt.polar(angles,data,color='k',marker='o')
plt.xticks(angles,labels)
plt.title('数据分析师综合评级')
```

### 棉棒图

```python
plt.rcParams['font.sans-serif']='Simhei'
plt.rcParams['axes.unicode_minus']=False

x = np.linspace(0.5,2*np.pi,30)
y = np.random.randn(30)
y = [float('%3.1f'%j )for j in y]
for i,j in zip(x,y):
    plt.text(i,j,j+1,ha='center',va='bottom')

plt.stem(x,y,linefmt='k-.',markerfmt='ko',basefmt='k-')
```

```python
x = np.linspace(0.1,0.6,6)
y = np.exp(x)

error = 0.05+0.15*x
lower_error = error
upper_error = 0.3*error
error_limit = [lower_error,upper_error]

plt.errorbar(x,y,fmt='ko-',yerr=error_limit,xerr=0.02,ecolor='k',elinewidth=0.5,lw=1,capsize=2,barsabove=True,errorevery=1)
```

```python
import matplotlib.pyplot as plt
#plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
#plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
# 有中文出现的情况，需要u'内容'
import numpy as np
from matplotlib.sankey import Sankey
sankey = Sankey(scale=0.01, unit='%', format='%.0f', gap=0.25, radius=0.1,
       shoulder=0.03, offset=0.15, head_angle=100, margin=0.3, tolerance=1e-06)
sankey.add(flows=[97.4, 1.2, 1.4, -19.8, -22.5, -10.0, -12.1, -9.0,-17.5,-8.9],
    labels=['a', '', '', '', '', '', '', '', '', ''],
    orientations=[0, -1, 1, 1, 1, 1, 0, -1,-1,-1],
      pathlengths=[0.5, 0.5, 0.5, 0.25, 0.25, 0.6, 0.25, 0.25,
                   0.25,0.25])
diagrams = sankey.finish()
diagrams[0].patch.set_facecolor('white')
plt.show()
```
