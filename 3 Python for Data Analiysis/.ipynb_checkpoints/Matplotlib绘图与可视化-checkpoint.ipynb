```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
ax3 = plt.plot(np.random.randn(50).cumsum(),'k--')
```

```python
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

### 颜色标记和线类型

```python
plt.plot(np.random.randn(30).cumsum(),'ko--')
```

```python
plt.plot(np.random.randn(30).cumsum(),color='k',linestyle='dashed',marker='o')
```

```python
data = np.random.randn(30).cumsum()
```

```python
plt.plot(data,'k--',label='Default')
plt.plot(data,'k-',drawstyle='steps-post',label='steps-post')
plt.legend(loc='best')
```

### 刻度标签和图例

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(list('ABCDE'),fontsize='small')
ax.set_title('My First')
ax.set_xlabel('Stages')
ax.plot(np.random.randn(1000).cumsum(),'k',lw=0.8)
```

### 使用pandas绘图

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.plot(subplots=True,sharex=True,rot=0,sharey=True)
```

```python
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```
