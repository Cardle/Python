```python
datetime.now().day
```

```python
datetime.now().weekday()+1
```

```python
datetime.now().time() #只显示时间
```

```python
datetime.now().strftime('%F %W %w')
```

```python
cha.seconds
```

```python
cha.seconds/3600
```

```python
date = datetime(2019,4,7,18,18)
date + timedelta(days = 1) #忘后推移一天
```

```python
date + timedelta(seconds = 60) #忘后推移60秒
```

```python
date - timedelta(days = 1) #忘前推移一天
```

```python
date - timedelta(seconds = 60) #忘前推移60秒
```

```python
date + Day(1)
```

```python
date + Hour(2)
```

```python
date + Minute(10)
```

### Matplotlib绘图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel(r'C:\Users\Nikola\Desktop\Python\Matplotlib.xlsx')
df1
```

```python
%matplotlib inline
```

```python
plt.rcParams['font.sans-serif']='SimHei' #解决中文乱码问题
plt.rcParams['axes.unicode_minus']=False #解决负号无法显示的问题
%config InlineBackend.figure_format = 'svg' #将图表设置为svg格式以保证更为清晰的显示
```

```python
#利用add_subplot（）函数建立坐标系
fig = plt.figure() #建立画布
ax1 = fig.add_subplot(2,2,1) #建立2x2的坐标系
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
x = np.arange(6)
y = np.arange(6)
```

```python
# 利用subplot2grid()函数建立坐标系
plt.subplot2grid((2,2),(0,0))#利用此函数时不需要先建立画布
plt.plot(x,y)
plt.subplot2grid((2,2),(0,1))#具体的绘图代码需要跟在建立坐标系的语句后面
plt.bar(x,y)
plt.subplot2grid((2,2),(1,0))
plt.bar(x,y)
plt.subplot2grid((2,2),(1,1))
plt.plot(x,y)
```

```python
#利用plt.subplot()函数建立坐标系
plt.subplot(221) 
plt.plot(x,y)
plt.subplot(224)
plt.bar(x,y)
```

```python
#利用plt.subplots()函数建立坐标系，区别在于以上函数一次只能返回一个坐标系，而它可以一次返回多个坐标系
```

```python
fig, axes = plt.subplots(2,2)
```

```python
axes[0,0].plot(x,y) #运行后图表不自动显示
```

```python
axes[1,1].bar(x,y) #运行后图表不自动显示
```

```python
#利用add_subplot（）函数建立坐标系
fig = plt.figure() #建立画布
ax1 = fig.add_subplot(1,1,1) #建立1x1的坐标系
plt.plot(df1['人数']) #以df的列绘制折线图

plt.xlabel('月份',labelpad = 5, fontsize = 'xx-large',color = 'k',fontweight = 'bold') 
#标签名称，与x轴的距离，标签大小，标签颜色，标签加粗
plt.ylabel('注册人数',labelpad = 5) #设置x轴标签与Y轴标签，labelpad参数为标签到xy轴的距离

# plt.xticks(np.arange(0,12),[str(i)+"月" for i in range(1,13)]) #设置x轴刻度标签以及刻度值
# plt.yticks(np.arange(0,2000,400),[str(j)+"人" for j in range(0,2000,400)]) #设置y轴客户标签以及刻度值
#plt.yticks([]) #隐藏Y轴刻度以及客户值
```
