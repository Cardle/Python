### 绘制折线图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
plt.rcParams['font.sans-serif']='SimHei' #解决中文乱码问题
plt.rcParams['axes.unicode_minus']=False #解决负号无法显示的问题
%config InlineBackend.figure_format = 'svg' #将图表设置为svg格式以保证更为清晰的显示
```

```python
%matplotlib inline
```

```python
x = np.array([str(i)+"月" for i in range(1,13)] )#取1-9月作为x轴
y = np.array(random.sample(range(500,1000),12)) # 取一组随机整数值作为y值
print(x)
print(y)
```

```python
# 建立一个8x4的画布
fig = plt.figure(figsize=(8,4))

# 建立一个坐标系
plt.subplot(111)

# 隐藏坐标系上中右边框
plt.subplot(111).spines['top'].set_visible(False)
plt.subplot(111).spines['left'].set_visible(False)
plt.subplot(111).spines['right'].set_visible(False)

# 绘制折线图
plt.plot(x,y,color='k',linestyle='dashed',linewidth=1, marker='o',markersize=4, label='折线图') 

# 设置坐标轴标题
plt.xlabel('月份',fontweight='bold',fontsize=11)
plt.ylabel('人数')

# 设置轴刻度线,隐藏左边的刻度线，隐藏左边的刻度标签
plt.tick_params(axis='x',which='major',direction='in',left=False, labelleft=False)


# 添加图表标题
plt.title('2019年月度用户注册量',loc='right',pad=10, fontdict=dict(fontsize=11,fontweight='bold'))

# 添加数据标签
for a,b in zip(x,y):
    plt.text(a,b,b,ha="center",va='bottom')

# 设置网格线
# plt.grid()
plt.grid('false')

# 设置图例：调用显示出plot中的label值
plt.legend()

# 设置隐藏坐标轴刻度线
# plt.axis('off')

# 保存图表到本地
plt.savefig(r'C:\Users\Nikola\Desktop\Python\折线图.jpg')
```

### 绘制柱形图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
%matplotlib inline
```

```python
plt.rcParams['font.sans-serif']='Simhei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
```

```python
x = np.array(['东区','南区','西区','北区'])
y1 = np.array([1000,3000,2000,4000])
y2 = np.array(random.sample(range(3000,8000),4))
print(x)
print(y1)
print(y2)
```

```python
#绘制常规柱形图
plt.figure(figsize=(5,4))
plt.subplot(111)
plt.bar(x,y1,width=0.5,align='center',label='任务量',color=(0.004,0.722,0.667))
plt.title('全国分区任务量',loc='center')
plt.xlabel('区域')
plt.ylabel('任务量')
for a,b in zip(x,y1):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
plt.legend()
```

```python
#绘制簇状柱形图
x = np.array([1,2,3,4])
y1 = np.array([1000,3000,2000,4000])
y2 = np.array([3000,2000,4000,5000])
plt.figure(figsize=(5,4))
plt.subplot(111)
plt.subplot(111).spines['top'].set_visible(False)
plt.subplot(111).spines['left'].set_visible(False)
plt.subplot(111).spines['right'].set_visible(False)
plt.bar(x,y1,width=0.3,label='任务量',color=(0.004,0.722,0.667))
plt.bar(x+0.3,y2,width=0.3,label='完成量',color=(0.949,0.784,0.059))
plt.title('全国分区任务量与完成量',loc='center')
plt.xlabel('区域')
plt.ylabel('任务量')
plt.tick_params(left=False, labelleft=False)
plt.xticks(x+0.15,['东区','南区','西区','北区'])
for a,b in zip(x,y1):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
for a,b in zip(x+0.3,y2):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
plt.legend(ncol=2)
plt.savefig(r'C:\Users\Nikola\Desktop\Python\簇状图.svg')
```

```python
#绘制堆积柱形图
x = np.array([1,2,3,4])
y1 = np.array([3000,2000,1000,4000])
y2 = np.array([1000,5000,3000,4000])
plt.figure(figsize=(5,4))
plt.subplot(111)
plt.subplot(111).spines['top'].set_visible(False)
plt.subplot(111).spines['left'].set_visible(False)
plt.subplot(111).spines['right'].set_visible(False)
plt.bar(x,y1,width=0.3,label='任务量',color=(0.004,0.722,0.667))
plt.bar(x,y2,width=0.3,label='完成量',color=(0.949,0.784,0.059),bottom=y1)
plt.title('全国分区任务量与完成量',loc='center')
plt.xlabel('区域')
plt.ylabel('任务量')
plt.tick_params(left=False, labelleft=False)
plt.xticks(x,['东区','南区','西区','北区'])
for a,b in zip(x,y1):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
for a,b in zip(x,y1+y2):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
plt.legend(ncol=2)
plt.savefig(r'C:\Users\Nikola\Desktop\Python\堆积柱形图.svg')
```

### 绘制条形图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
%matplotlib inline
```

```python
plt.rcParams['font.sans-serif']='Simhei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
```

```python
plt.subplot(111)
x = np.array(['东区','南区','西区','北区'])
y = np.array([8566, 6482, 5335, 7310])
plt.barh(x,height=0.5,width=y,align='center',color=(0.004,0.722,0.667))
plt.title('全国各区任务量',loc='center')
plt.xlim(1,10000)
for a,b in zip(x,y):
    plt.text(b,a,b,ha='left',va='center',fontsize=11)
plt.savefig(r'C:\Users\Nikola\Desktop\Python\条形图.svg')
```
