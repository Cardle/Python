### 绘制折线图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
plt.rcParams['font.sans-serif']='SimHei' #解决中文乱码问题
plt.rcParams['axes.unicode_minus']=False #解决负号无法显示的问题
%config InlineBackend.figure_format = 'svg' #将图表设置为svg格式以保证更为清晰的显示
```

```python
%matplotlib inline
```

```python
x = np.array([str(i)+"月" for i in range(1,13)] )#取1-9月作为x轴
y = np.array(random.sample(range(500,1000),12)) # 取一组随机整数值作为y值
print(x)
print(y)
```

```python
# 建立一个8x4的画布
fig = plt.figure(figsize=(8,4))

# 建立一个坐标系
plt.subplot(111)

# 隐藏坐标系上中右边框
plt.subplot(111).spines['top'].set_visible(False)
plt.subplot(111).spines['left'].set_visible(False)
plt.subplot(111).spines['right'].set_visible(False)

# 绘制折线图
plt.plot(x,y,color='k',linestyle='dashed',linewidth=1, marker='o',markersize=4, label='折线图') 

# 设置坐标轴标题
plt.xlabel('月份',fontweight='bold',fontsize=11)
plt.ylabel('人数')

# 设置轴刻度线,隐藏左边的刻度线，隐藏左边的刻度标签
plt.tick_params(axis='x',which='major',direction='in',left=False, labelleft=False)


# 添加图表标题
plt.title('2019年月度用户注册量',loc='right',pad=10, fontdict=dict(fontsize=11,fontweight='bold'))

# 添加数据标签
for a,b in zip(x,y):
    plt.text(a,b,b,ha="center",va='bottom')

# 设置网格线
# plt.grid()
plt.grid('False')

# 设置图例：调用显示出plot中的label值
plt.legend()

# 设置隐藏坐标轴刻度线
# plt.axis('off')

# 保存图表到本地
plt.savefig(r'C:\Users\Nikola\Desktop\Python\折线图.jpg')
```
